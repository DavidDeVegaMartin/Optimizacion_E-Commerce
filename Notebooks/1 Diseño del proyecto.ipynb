# 1 DISEÑO DEL PROYECTO

# 1.0 Introduccion

El E-commerce ACME Cosmetic esta viviendo una situacion de estancamiento y ralentizacion de las ventas.

Por ello encarga al equipo de data science recien creado el anailisis de sus datos transaccionales e implementar acciones CRO personalizacadas a su situacion.

# 1.1 Objetivo

**Analizar los datos transaccionales para intentar potenciales acciones CRO que incrementen las visitas, conversiones y ticket medio, y por tanto incrementar la facturacion global del e-commerce.**

Productos a diseñar:

* Segmentacion RFM.

* Sistema de recomendación.

# 1.2 KPIs y Entidades de datos

* KPIS relevantes:

  - Visitas.
 
  -  Tasa de Conversion.
 
  - Frecuencia de compra.
 
  - Ticket medio.
 
  - Tasa abandono carrito.
 
  - LTV (LifeTime Value).
 

* Entidades:

  - Usuarios , persona que navega por la web.
 
  - Clientes, usuario que formaliza una compra.
 
  - Sesiones.
 
  - Eventos.
 
  - Productos.
 
# 1.3 Preguntas Semilla

* Sobre Customer journey

    - ¿Cuál es el flujo habitual que sigue un usuario desde que entra al sitio hasta concretar una compra?

    - En promedio, ¿cuántos artículos visualiza, añade al carrito, abandona o efectivamente compra un usuario durante una sesión?

    - ¿Cómo han evolucionado estos comportamientos en los últimos meses?

* Sobre los clientes:

    - ¿Cuál es el número medio de productos adquiridos por cada cliente?

    - ¿Qué importe total gasta, en promedio, cada comprador?

    - ¿Podemos identificar a los clientes con mayor valor o frecuencia de compra, y ofrecerles un tratamiento diferenciado?

    - ¿Con qué frecuencia los clientes vuelven a realizar compras tras su primera transacción?

    - ¿Cuál es el valor promedio del ciclo de vida de un cliente (LTV)?

    - ¿Es viable implementar estrategias de marketing personalizadas basadas en el valor estimado de cada cliente?

* Sobre los productos:

    - ¿Qué artículos tienen el mayor volumen de ventas?

    - ¿Existen productos que no han sido adquiridos en absoluto?

    - ¿Se observa alguna correlación entre el precio de un artículo y la cantidad de unidades vendidas?

    - ¿Qué productos reciben muchas visitas pero no logran convertir en ventas?

    - ¿Hay artículos que son eliminados con frecuencia del carrito de compras?

    - ¿Podríamos ofrecer recomendaciones personalizadas de productos a cada cliente en función de su comportamiento?
